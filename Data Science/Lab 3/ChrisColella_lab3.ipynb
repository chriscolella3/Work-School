## Exercise 1 (8 points)
Clean the data in the DataFrame **Energy**. The detailed steps are as follows.  
  
#### Steps:   
1. For all the countries which have missing data (e.g., data with '...'), represent them as np.NaN values. 
2. Rename the following list of countries: 
        "Republic of Korea": "South Korea",
        "United States of America": "United States",
        "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
        "China, Hong Kong Special Administrative Region": "Hong Kong"


```python
energy.replace("...",np.NaN, inplace=True)
energy['Country Name'].replace({"Republic of Korea": "South Korea",
 "United States of America": "United States",
 "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
 "China, Hong Kong Special Administrative Region": "Hong Kong"
}, inplace=True)

energy[energy['Country Name'] == "Hong Kong"]
```

####  Execute the following statements
1. Because of '...' in the original data, the columns 'Energy Supply' and 'Energy Supply per capita' are recognized as strings. Use **astype** to convert them to be floating point number. 
2. "Energy Supply" is in petajoules but "Energy Supply per capita" is in gigajoules. Convert the "Energy Supply" to gigajoules (there are 1,000,000 gigajoules in a petajoule). 

```python
energy['Energy Supply'] = energy['Energy Supply'].astype(float)
energy['Energy Supply per capita'] = energy['Energy Supply per capita'].astype(float)
energy["Energy Supply"] = energy["Energy Supply"]*1000000
```

## Exercise 2 (8 points)
Process the DataFrame **gdp**. The detailed steps are as follows.  
  
#### Steps:   
1. Rename the following list of countries in "Country Name": 
        "Korea, Rep.": "South Korea", 
        "Iran, Islamic Rep.": "Iran",
        "Hong Kong SAR, China": "Hong Kong"
2. Copy the column "Country Name" and the last 5 years (2013-2017) of GDP data to the DataFrame **GDP**.


```python
gdp['Country Name'].replace({ "Korea, Rep.": "South Korea", 
 "Iran, Islamic Rep.": "Iran",
 "Hong Kong SAR, China": "Hong Kong"
}, inplace=True)

GDP = gdp[['Country Name', '2013', '2014', '2015', '2016', '2017']]
GDP
```

## Exercise 3 (8 points)
Join the dataset **energy** and **GDP**. Create a new DataFrame **energyGDP** which select the top 10 countries in energy supply of the joined dataset. Set the column "Country Name" to be the index of "energyGDP".
  
#### Steps:   
1. Merge **energy** and **GDP** with the key "Country Name".
2. Sort the merged dataset by the values in the column 'Energy Supply' in descending order.
3. Select the first 10 rows of the sorted dataset and assign it to **energyGDP**.
4. Set the column "Country Name" to be the index of "energyGDP".

```python
energyGDP = energy.merge(GDP, on="Country Name")

energyGDP.sort_values(by='Energy Supply', ascending=False)[:10]

energyGDP.set_index("Country Name", inplace=True)

energyGDP
```

## Exercise 4 (8 points)
Based on the DataFrame **GDP**, create a new DataFrame with two levels of index and one column. The outer index is the country name and the inner index is the year. The value of the column is the corresponding GDP. Furthermore, calculate the average GDP of each country from 2013 to 2017.

#### Steps:

1. Using pandas.melt, create a temporary DataFrame **tmp1** to merge all the columns of years in **GDP** into one. Set the 'country name' column as indicator and other columns as data values.
2. Create a new DataFrame **newGDP** and set the column "Country Name" and "variable" to be the index of it.
3. Apply **mean** method and set the **level** to be 'Country Name'.

```python
tmp1 = pd.melt(GDP, id_vars =['Country Name'])
newGDP = tmp1.set_index(["Country Name", "variable"])
newGDP.mean(level="Country Name")
```
